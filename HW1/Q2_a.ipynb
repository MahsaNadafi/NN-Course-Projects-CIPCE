```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Points
np.random.seed(1)
x1 = np.random.normal(1, 0.3, 100)
y1 = np.random.normal(1, 0.3, 100)
x2 = np.random.normal(-1, 0.3, 100)
y2 = np.random.normal(-1, 0.3, 100)

# Plot Points
plt.scatter(x1, y1, marker="o", label='~N(1,0.3)', color = 'blue')
plt.scatter(x2, y2, marker="o", label='~N(-1,0.3)', color = 'red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Generated Points')
plt.legend(loc= 'upper left')
plt.show()
```

```python
# Data matrix
data = np.zeros((200, 2))
data_label = np.zeros((200, 1)) # target (t)

# 1st group
data[0:100, 0] = x1
data[0:100, 1] = y1
data_label[0:100] = 1  

# 2nd group
data[100:200, 0] = x2
data[100:200, 1] = y2
data_label[100:200] = -1  

# learning parameter initialization
w = [0.1] * 2 
s = [0] * 2
b = 0.1
lr = 0.01 # learning rate
Cost = []
totalCost = 10
TotalCost = []
j= 0 # iteration
while j < 10 and totalCost > 3:
  j = j + 1
  totalCost = 0
  for i in range(0, 200):
    s = data[i, 0:2]
    net = np.dot(w, s) + b
    t = data_label[i]
    # updating parameters
    w = w + lr * (t - net) * s
    b = b + lr * (t - net)
    # computing cost
    cost = 0.5 * (t - net) ** 2
    totalCost = totalCost + cost
    Cost.append(cost)
  TotalCost.append(totalCost)
  print(totalCost)

print(len(Cost))

plt.plot(x1, y1, 'o', label='1st group', color= 'blue')
plt.plot(x2, y2, 'o', label='2nd group', color= 'red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('two group of randomly generated points')
plt.legend(loc= 'upper left')
xmin, xmax = -2, 2
ymin, ymax = -2, 2
x = np.linspace(xmin, xmax, 20)
y = (-1 / w[1]) * (w[0] * x + b)
plt.plot(x, y)
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.show()
# cost curve
plt.plot(TotalCost)
plt.xlabel('iteration')
plt.ylabel('error')
plt.title('0.5 * (t - net) ** 2')
plt.show()
```
