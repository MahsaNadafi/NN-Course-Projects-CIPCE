```python
import numpy as np

# 2 bit Inputs A and B
A= np.array([1,0])
B= np.array([1,0])

# Define the threshold activation function
T= 0
b= -2 #bias
def linear_threshold_gate(dot: int, T: float) -> int:
    '''Returns the binary threshold output'''
    if dot >= T:
        return 1
    else:
        return 0
```

```python
# Weight vector for and operation
W_and= np.array([1,1])
# Structring xor operation with Mcculloch-Pitts Neuron
# Weight vector for and not operation
W_andnot= np.array([2,-1])
W_or= np.array([2,2])
```

```python
# Reshape inputs for Mcculloch-Pitts Neuron
A0_B0= np.concatenate(([A[1]] , [B[1]]),axis= 0)
# Compute Output
A0_and_B0= A0_B0 @ W_and + b 
# Compute the output based on the threshold value
C0 = linear_threshold_gate(A0_and_B0, T) # Output C3- Neuron Number 1
print(f'C0: {C0}')
```

```python
# Reshape inputs for Mcculloch-Pitts Neuron
A0_B1= np.concatenate(([A[1]] , [B[0]]), axis= 0)
A1_B0= np.concatenate(([A[0]] , [B[1]]), axis= 0)

# Compute Output
A0andB1= A0_B1 @ W_and + b # A0 and B1
a0andb1 = linear_threshold_gate(A0andB1, T) # - Neuron Number 2
A1andB0= A1_B0 @ W_and + b # A1 and B0
a1andb0 = linear_threshold_gate(A1andB0, T) # - Neuron Number 3

Z1= np.concatenate(([a0andb1], [a1andb0]), axis= 0) @ W_andnot + b # A0andB1_andnot_A1andB0 
z1= linear_threshold_gate(Z1, T) # - Neuron Number 4
Z2= np.concatenate(([a1andb0], [a0andb1]), axis= 0) @ W_andnot + b # A0andB1_andnot_A1andB0
z2= linear_threshold_gate(Z2, T) # - Neuron Number 5
Z1_or_Z2= np.concatenate(([z1], [z2]), axis= 0) @ W_or + b # Output C2

# Compute the output based on the threshold value
C1 = linear_threshold_gate(Z1_or_Z2, T) # - Neuron Number 6
print(f'C1: {C1}')
```

```python
# Reshape inputs for Mcculloch-Pitts Neuron
A1_B1= np.concatenate(([A[0]] , [B[0]]), axis= 0)
# Compute Output
A1andB1= A1_B1 @ W_and + b
a1andb1 = linear_threshold_gate(A1andB1, T) # - Neuron Number 7
A0andB1_and_A1andB0= np.concatenate(([a1andb1], [a1andb0]), axis= 0) @ W_and + b
a0andb1_and_a1andb0 = linear_threshold_gate(A0andB1_and_A1andB0, T)  # - Neuron Number 8

Z3= np.concatenate(([a1andb1], [a0andb1_and_a1andb0]), axis= 0) @ W_andnot + b # (A0andB1_and_A1andB0)xor(A1andB1)
z3= linear_threshold_gate(Z3, T) # - Neuron Number 9
Z4= np.concatenate(([a0andb1_and_a1andb0], [a1andb1]), axis= 0) @ W_andnot + b # (A0andB1_and_A1andB0)xor(A1andB1)
z4= linear_threshold_gate(Z4, T) # - Neuron Number 10
Z3_or_Z4= np.concatenate(([z3], [z4]), axis= 0) @ W_or + b # Output C1 - Neuron Number 9

# Compute the output based on the threshold value
C2 = linear_threshold_gate(Z3_or_Z4, T)
print(f'C2: {C2}')
```

```python
A1andB1__and__A0andB1_and_A1andB0= np.concatenate(([a1andb1], [a0andb1_and_a1andb0]), axis= 0) @ W_and + b # (A1andB1)and(A0andB1_and_A1andB0)
# Compute the output based on the threshold value
C3 = linear_threshold_gate(A1andB1__and__A0andB1_and_A1andB0, T) # - Neuron Number 10
print(f'C3: {C3}')
```
